```python
import requests
from bs4 import BeautifulSoup
import tkinter as tk
from tkinter import messagebox
import matplotlib.pyplot as plt

# Function to get weather data for a given city
def get_weather(city):
    # Creating URL and requests instance
    url = "https://www.google.com/search?q=" + "weather" + city
    html = requests.get(url).content

    # Getting raw data
    soup = BeautifulSoup(html, 'html.parser')
    temp = soup.find('div', attrs={'class': 'BNeawe iBp4i AP7Wnd'}).text
    str_data = soup.find('div', attrs={'class': 'BNeawe tAd8D AP7Wnd'}).text

    # Formatting data
    data = str_data.split('\n')
    time = data[0]
    sky = data[1]

    # Getting other required data
    list_div = soup.findAll('div', attrs={'class': 'BNeawe s3v9rd AP7Wnd'})
    str_div = list_div[5].text
    pos = str_div.find('Wind')
    other_data = str_div[pos:]

    return temp, time, sky, other_data

# Function to handle button click event
def get_weather_info():
    city = entry_city.get()
    if city:
        try:
            temp, time, sky, other_data = get_weather(city)
            result_text = f"Temperature: {temp}\nTime: {time}\nSky Description: {sky}\n{other_data}"
            messagebox.showinfo("Weather Information", result_text)
            
            # Create a pie chart
            labels = ['Temperature', 'Time', 'Sky Description', 'Other Data']
            sizes = [1, 1, 1, 1]  # Equal sizes for demonstration
            plt.pie(sizes, labels=labels, autopct='%1.1f%%')
            plt.title('Weather Information')
            plt.show()
        except Exception as e:
            messagebox.showerror("Error", f"Failed to retrieve weather information. {e}")
    else:
        messagebox.showwarning("Warning", "Please enter a city name.")

# Create GUI window
window = tk.Tk()
window.title("Weather Information")

# Create and place widgets
label_city = tk.Label(window, text="Enter city name:")
label_city.grid(row=0, column=0, padx=10, pady=10)

entry_city = tk.Entry(window)
entry_city.grid(row=0, column=1, padx=10, pady=10)

button_get_weather = tk.Button(window, text="Get Weather", command=get_weather_info)
button_get_weather.grid(row=1, columnspan=2, padx=10, pady=10)

# Run the GUI event loop
window.mainloop()
```
